## gen_greyscale_shear

```python
image = Image.open('Austausch_IPD_Optimierung/gen_greyscale_shear/shear.png')
print('size: ', image.size)
display(image)
```

```python
pixel_array = image.convert("L").load()

min_value = pixel_array[0, 0]
max_value = pixel_array[0, 0]

for y in range(image.size[1]):
    for x in range(image.size[0]):
        pixel_value = pixel_array[x, y]
        if pixel_value < min_value:
            min_value = pixel_value
        if pixel_value > max_value:
            max_value = pixel_value

print(f"pixel value: [{min_value}, {max_value}]")
```

## Lengths

```python
data = pd.read_csv('Austausch_IPD_Optimierung/u_spring/U_spring_doubledome.csv')
data = pd.read_csv('2023_06/04_Spring_Elongation/U_spring_L_shape.csv')
print('u: ', data['u'].min(), data['u'].max())
```

```python
grippers_coordinates = np.load('Austausch_IPD_Optimierung/spring_coords/spring_pnt_of_attack_coords.npy')[:2]
pair_lengths = []
for i in range(len(data)):
    for j in range(i + 1, len(data)):
        if i != j:
            if (grippers_coordinates[0][i] == grippers_coordinates[0][j]) and (
                    abs(grippers_coordinates[1][i] - grippers_coordinates[1][j]) == 460):
                pair_lengths.append(data.iloc[i]['u'] + data.iloc[j]['u'])
            if (grippers_coordinates[1][i] == grippers_coordinates[1][j]) and (
                    abs(grippers_coordinates[0][i] - grippers_coordinates[0][j]) == 300):
                pair_lengths.append(data.iloc[i]['u'] + data.iloc[j]['u'])

min(pair_lengths), max(pair_lengths)
```
